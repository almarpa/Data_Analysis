Ejercicio 4. Escribe una función en Python que reciba como parámetros una gramática incontextual y una oración a analizar. La función mostrará el árbol sintáctico de la oración, en caso de que ésta sea correcta según la gramática dada. Si la oración no es correcta, mostrará el mensaje "Oración no reconocida". En caso de ser ambigua, mostrará el mensaje "Oración ambigua" y visualizará todos los árboles sintácticos posibles.

Utiliza el analizador sintáctico *RecursiveDescentParser* incluido en la distribución nltk. 
Prueba con la gramática dada, *gramatica.cfg*, y con las siguientes oraciones:

*"¿ Qué mares bañan Andalucía ?"*

*"¿ El río Ebro pasa por el País_vasco ?"*

*"¿ Cuántos ríos desembocan en el mar Mediterráneo ?"*

*"El río Ebro por pasa el País_vasco ."*


```python
def demo(oracion):

    #Escribe aquí tu código Python y ejecútalo
    from nltk import parse, CFG

    grammar = CFG.fromstring(
        """
    S -> S_dec
    S -> S_int
    S_dec -> SN SV Fp
    S_int -> Fia SN SV Fit
    S_int -> Fia SN_int SV Fit
    SN -> NOMINAL 
    SN -> NOMINAL AQ 
    SN -> DA NOMINAL
    SN -> DA NOMINAL AQ
    SN_int -> DT NOMINAL 
    SP -> SPS SN
    NOMINAL -> NC 
    NOMINAL -> NP
    NOMINAL -> NC NP
    SV -> VMI SN 
    SV -> VMI SP
    VMI -> 'pasa'
    VMI -> 'desembocan' 
    VMI -> 'bañan'
    Fia -> '¿'
    Fit -> '?'
    Fp -> '.'
    SPS -> 'por' 
    SPS -> 'en'
    SPS -> 'con'
    DA -> 'el'
    DA -> 'El'
    DT -> 'Qué' 
    DT -> 'Cuántos'
    NC -> 'comunidad' 
    NC -> 'mar' 
    NC -> 'mares' 
    NC -> 'río' 
    NC -> 'ríos'
    AQ -> 'valenciana' 
    AQ -> 'gallega' 
    NP -> 'Ebro' 
    NP -> 'País_vasco' 
    NP -> 'Andalucía' 
    NP -> 'Mediterráneo'
    SN -> DA NOMINAL CN
    SPV -> SPS ADV NCV
    SPN -> SPS ADV NCN
    SV -> VMI SP CV
    CN -> SPN
    CV -> SPV
    ADV -> 'más'
    SPS -> 'con'
    DA -> 'la'
    NC -> 'comunidad'
    NCN -> 'habitantes'
    NCV -> 'caudal' 
    """
    )

    
    for prod in grammar.productions():
        print(prod)

    sent = oracion.split()
    parser = parse.RecursiveDescentParser(grammar, trace=2)
    for p in parser.parse(sent):
        print(p)

demo("¿ Qué río pasa por la comunidad con más habitantes ?")

```
