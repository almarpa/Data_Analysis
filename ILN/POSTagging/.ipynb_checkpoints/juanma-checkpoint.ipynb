```python
from collections import OrderedDict

tags = brown.tagged_words(tagset = 'universal')
aparitions = {}
counts = []

for tag in tags:
    if tag[0] not in aparitions:
        aparitions[tag[0]] = [tag[1]]
    elif tag[0] in aparitions and tag[1] not in aparitions[tag[0]]:
        aparitions[tag[0]].append(tag[1])
 


for key,value in aparitions.items():
    aparitions[key] = len(value)

aparitions = OrderedDict(sorted(aparitions.items(), key=lambda x: x[0]))
aparitions = OrderedDict(sorted(aparitions.items(), key=lambda x: x[1], reverse =True))
print("{:<10} {:<10}".format('Palabra', 'Etiquetas'))
for item in aparitions.items():
    print("{:<10} {:<10}".format(item[0], item[1]))
```
